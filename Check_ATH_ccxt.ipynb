```python
import ccxt
from pprint import pprint
import pandas as pd
import time
import datetime
from datetime import timezone
import timeit
```

# Change "Exchange_Name"

```python
exchange_name = 'ftx' # initial exchange
t_frame = '1d' # 1-day timeframe, usually from 1-minute to 1-week depending on the exchange
symbol = '' # initial symbol

#Start_date = datetime.date(2021,1,1) # Start_date Year/ Month/ Date
#####
#startdate = time.mktime(Start_date.timetuple())
```

```python
# Get our Exchange
try:
    exchange = getattr (ccxt, exchange_name) ()
except AttributeError:
    print('-'*36,' ERROR ','-'*35)
    print('Exchange "{}" not found. Please check the exchange is supported.'.format(exch))
    print('-'*80)
```

# Change Date

```python
startdate = exchange.parse8601('2020-01-01T00:00:00') # Start_date Year/ Month/ Date
startdate
```

```python
# Check if the symbol is available on the Exchange
exchange.load_markets()
if symbol not in exchange.symbols:
    print('-'*36,' ERROR ','-'*35)
    print('The requested symbol ({}) is not available from {}\n'.format(symbol,exchange_name))
    print('Available symbols are:')
    for key in exchange.symbols:
        print('  - ' + key)
    print('-'*80)
```

```python
# Check if the symbol is available on the Exchange
exchange.load_markets()
if symbol not in exchange.symbols:
    print('-'*36,' ERROR ','-'*35)
    print('The requested symbol ({}) is not available from {}\n'.format(symbol,exchange_name))
    print('Available symbols are:')
    symbol_list = []
    for key in exchange.symbols:
        symbol_list.append(key)
    print('-'*80)
    print("Create List from {}".format(exchange_name.upper()))
```

```python
def get_ath(lisa):
  for i in lisa:
    try:
      data = exchange.fetch_ohlcv(i, t_frame, startdate, limit= None)
      datadf = pd.DataFrame(data,columns=["Date", "Open", "High", "Low", "Close", "Vol"])
      if datadf.iloc[-1]["Close"] == datadf["Close"].max():
        print(i , "today is All-Time High")
      else :
        pass
    except:
      continue
```

```python
start = timeit.default_timer()
get_ath(symbol_list)
stop = timeit.default_timer()
print( 'Running time: ', stop-start )
```
