```python
ylist_symbol = []

# -*- coding: utf-8 -*-
from bs4 import BeautifulSoup
import requests
urls=["https://finance.yahoo.com/cryptocurrencies?.tsrc=fin-srch&count=100&offset=0", 'https://finance.yahoo.com/cryptocurrencies?count=100&offset=100', 'https://finance.yahoo.com/cryptocurrencies?count=100&offset=200', 'https://finance.yahoo.com/cryptocurrencies?count=100&offset=300']
header = {
	'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_2) AppleWebKit/601.3.9 (KHTML, like Gecko) Version/9.0.2 Safari/601.3.9'
}

for ur in urls:
  response=requests.get(ur,headers=header)
  soup=BeautifulSoup(response.content, 'lxml')
  for item in soup.select('.simpTblRow'):
	  ylist_symbol.append(item.select('[aria-label=Symbol]')[0].get_text())
```

```python
today_day = "{0}-{1}-{2}".format(datetime.now().year, datetime.now().month, datetime.now().day)
print("Today :", today_day)
```

```python
for symbols in ylist_symbol:
  try:
    data = yf.download(tickers=symbols ,start = today_day, interval="1h")
    if data["Open"][-1] > data["Open"][0]*1.03:
      print(symbols, "diff > 3%")
    else:
      pass
  except: 
    continue
```
